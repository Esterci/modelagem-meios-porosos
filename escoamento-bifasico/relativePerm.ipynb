```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Swi = 0.1
Sor = 0.15
krwMax = 0.5
a = 8
kroMax = 1
b = 3
mu_w = 1.0e-03
mu_o = 20.0 * mu_w
K = 1e-12

def relativeSw(Sw):
    return (Sw - Swi) / (1.0 - Swi - Sor)

def correct_krw(Sw):
    return krwMax * (Sw)**a

def correct_kro(Sw):
    return kroMax * (1 - Sw)**b

transmissibility = lambda K,krw,kro,mu_w,mu_o : K*((krw/mu_w)+(kro/mu_o))
```

```python
Sw = np.linspace(0.2,0.8,num=25,endpoint=True)

Swr = relativeSw(Sw)
kro = correct_kro(Swr)
krw = correct_krw(Swr)

plt.plot(Sw,kro,label="kro")
plt.plot(Sw,krw,label="krw")
plt.legend()
plt.grid()
plt.show()
```

```python
Tr = transmissibility(K,correct_krw(0.8),correct_kro(0.8),mu_w,mu_o)
Tr
```

```python
Vt = 1e-5

grad_p = Vt/Tr

grad_p
```
